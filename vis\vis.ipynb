```python
import numpy as np
import matplotlib as mpl
import pylab as pl
```

```python
#dati
xs=np.array([1.405, 1.988, 2.556, 3.03])
ys=np.array([0.0001125, 0.0001456, 0.0001775, 0.0002289])
```

```python
fig = pl.figure(figsize=(7,7))
ax= fig.add_subplot(111)
ax.plot(xs, ys, '.')
```

```python
def line(a,b,x):
  return a+b*x
```

```python
b=7.28*10**(-5)
a=2*10**(-7)

mptx=np.arange(1.3, 3.3, 0.1)
```

```python
mpty=line(a,b, mptx)
```

```python
yss=ys*1000
mptyy=mpty*1000
```

```python
fig = pl.figure(figsize=(5,5))
ax= fig.add_subplot(111)

ax.plot(xs, yss, '.', ms=8)
ax.plot(mptx,mptyy, 'r-', label='Ohm\'s law')
pl.xlabel("Voltage (V)")
pl.ylabel("Current (mA)")
ax.legend()
ax.grid(color='k', linestyle='--', alpha=0.2)
pl.xticks(ticks=[1.5,2,2.5,3])
```
